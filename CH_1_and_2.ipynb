```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Calculate pi:

history = dict()

for num_pts in [50001, 50002, 50003, 50004]:
    count_inside_circle = 0

    samples = []

    index = 0
    for x,y in zip(np.random.uniform(low=0, high=1, size=num_pts), np.random.uniform(low=0, high=1, size=num_pts)):
        dist = np.sqrt(x**2 + y**2)
        if dist <= 1:
            count_inside_circle += 1
        index += 1
        if index % 10 == 0:
            samples.append(4*count_inside_circle/index)
    
    print(f"π is (approx.) {4* np.round(count_inside_circle/num_pts, 4)}, Number Iterations:{num_pts}")

    history[num_pts] = samples
```

```python
colors = ['r', 'g', 'b', 'y']

index = 0
for x in [50001, 50002, 50003, 50004]:
    sample = history[x]

    plt.plot(sample, color=colors[index])
    index += 1
```

```python
# Integrate the function y=x**2 + 3*x in range [1,3] - should be 14(2/3)
# Now (min_y, max_y) = (4,18)

def ftn(x):
    return x**2 + 3*x


num_pts = 1000

pts = np.linspace(1,3,num_pts)
ftn_pts = ftn(pts)

random_pts_x = np.random.uniform(low=1, high=3, size=num_pts).reshape(num_pts,1)
random_pts_y = np.random.uniform(low=0, high=18, size=num_pts).reshape(num_pts,1)
random_pairs = np.concatenate([random_pts_x, random_pts_y], axis=1)


plt.plot(pts, ftn_pts)
plt.scatter(random_pairs[:,0], random_pairs[:,1])
```

```python
area_count = 0

for indx in range(random_pairs.shape[0]):
    x = random_pairs[indx,0]
    y = random_pairs[indx,1]

    if ftn(x) < y:
        area_count += 1

# Factor of 36 is because PDF = 1/36
print(f"Area estimate is {36 * area_count/random_pairs.shape[0]}")
```
